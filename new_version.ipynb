```python
"""
Мелкие скрипты для работы мое2й школы данные по муниципалитету
"""

import pandas as pd
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.utils import get_column_letter
from datetime import datetime
from openpyxl.chart import BarChart, Reference
from openpyxl.chart.label import DataLabelList
from openpyxl.styles import Alignment

import numpy as np
import os
```

```python
folder_data = 'z:\!!!ПРОЕКТЫ и МЕРОПРИЯТИЯ\ФГИС Моя Школа\Аналитика\Файлы'
path_end_folder = 'z:\!!!ПРОЕКТЫ и МЕРОПРИЯТИЯ\ФГИС Моя Школа\Аналитика'
```

```python
current_date = datetime.now()
formatted_date = current_date.strftime("%d_%m_%Y")
name_out_folder = current_date.strftime("%d_%m")
os.mkdir(f'{path_end_folder}/{name_out_folder}')
path_end_folder =f'{path_end_folder}/{name_out_folder}'

for file in os.listdir(folder_data):
    if (file.endswith('.xlsx') and not file.startswith('~$')):
        name_file = file.split('.xlsx')[0] # получаем название файла
        wb = openpyxl.load_workbook(f'{folder_data}/{file}')
        lst_sheet = [] # листы для обработки
        for sheet in  wb.sheetnames:
            if ('43' in sheet) or ('57' in sheet) or ('63' in sheet):
                lst_sheet.append(sheet)
            # удаляем лишние листы
        for sheet_name in wb.sheetnames[2:]:
            if not ('43' in sheet_name) or  not ('57' in sheet_name) or  not ('63' in sheet_name):
                del wb[sheet_name]              
                
        wb.create_sheet(title='43 Рейтинг активности',index=2)
        wb.create_sheet(title='Рейтинг муницип.',index=3)
        wb.create_sheet(title='Свод по муниципалитетам',index=4)
        df_43 = pd.read_excel(f'{folder_data}/{file}',sheet_name=lst_sheet[0]) # получаем рейтинг событий по муниципалитету
    #
        df_43.iloc[:,1] =df_43.iloc[:,1].astype(float)


        df_43.rename(columns={'Unnamed: 0':'Школа'},inplace=True)
        sort_column = df_43.columns[1]

        df_43.sort_values(by=sort_column,ascending=False,inplace=True)
        
        # округляем
        df_43[sort_column] = df_43[sort_column].apply(lambda x:round(x,1))

        for row in dataframe_to_rows(df_43,index=False,header=True):
            wb['43 Рейтинг активности'].append(row)
        # сохраняем по ширине колонок
        for column in wb['43 Рейтинг активности'].columns:
            max_length = 0
            column_name = get_column_letter(column[0].column)
            for cell in column:
                try:
                    if len(str(cell.value)) > max_length:
                        max_length = len(cell.value)
                except:
                    pass
            adjusted_width = (max_length + 2)
            wb['43 Рейтинг активности'].column_dimensions[column_name].width = adjusted_width
        
      
        #57 лист
        
        df_57 = pd.read_excel(f'{folder_data}/{file}',sheet_name=lst_sheet[1]) # получаем рейтинг событий по муниципалитету
    
        df_57.iloc[:,1] =df_57.iloc[:,1].astype(float)


        df_57.rename(columns={'Unnamed: 0':'Муниципалитет'},inplace=True)
        sort_column = df_57.columns[1]

        df_57.sort_values(by=sort_column,ascending=False,inplace=True)
        
        # округляем
        df_57[sort_column] = df_57[sort_column].apply(lambda x:round(x,1))

        for row in dataframe_to_rows(df_57,index=False,header=True):
            wb['Рейтинг муницип.'].append(row)
        # сохраняем по ширине колонок
        for column in wb['Рейтинг муницип.'].columns:
            max_length = 0
            column_name = get_column_letter(column[0].column)
            for cell in column:
                try:
                    if len(str(cell.value)) > max_length:
                        max_length = len(cell.value)
                except:
                    pass
            adjusted_width = (max_length + 2)
            wb['Рейтинг муницип.'].column_dimensions[column_name].width = adjusted_width
        
        
        # 63
        df_63 = pd.read_excel(f'{folder_data}/{file}',sheet_name=lst_sheet[2]) # получаем рейтинг событий по муниципалитету
        df_63 = df_63.drop([0])
        df_63 =df_63.apply(lambda x:pd.to_numeric(x,errors='ignore'))

        #df_63.rename(columns={'Unnamed: 0':'Муниципалитет'},inplace=True)
        df_63.columns= ['Муниципалитет','Законный представитель. Количество пользователей ФГИС «Моя школа»',
                        'Законный представитель. Количество активных пользователей ФГИС «Моя школа»',
                        'Обучающийся. Количество пользователей ФГИС «Моя школа»',
                        'Обучающийся. Количество активных пользователей ФГИС «Моя школа»',
                        'Педагогический работник. Количество пользователей ФГИС «Моя школа»',
                        'Педагогический работник. Количество активных пользователей ФГИС «Моя школа»']



        for row in dataframe_to_rows(df_63,index=False,header=True):
            wb['Свод по муниципалитетам'].append(row)
            
        wb['Свод по муниципалитетам'].column_dimensions['A'].width = 40
        wb['Свод по муниципалитетам'].column_dimensions['B'].width = 40
        wb['Свод по муниципалитетам']['B1'].alignment = Alignment(wrapText=True)
        
        wb['Свод по муниципалитетам'].column_dimensions['C'].width = 40
        wb['Свод по муниципалитетам']['C1'].alignment = Alignment(wrapText=True)
        
        wb['Свод по муниципалитетам'].column_dimensions['D'].width = 40
        wb['Свод по муниципалитетам']['D1'].alignment = Alignment(wrapText=True)
        
        wb['Свод по муниципалитетам'].column_dimensions['E'].width = 40
        wb['Свод по муниципалитетам']['E1'].alignment = Alignment(wrapText=True)
        
        wb['Свод по муниципалитетам'].column_dimensions['F'].width = 40
        wb['Свод по муниципалитетам']['F1'].alignment = Alignment(wrapText=True)
        
        wb['Свод по муниципалитетам'].column_dimensions['G'].width = 40
        wb['Свод по муниципалитетам']['G1'].alignment = Alignment(wrapText=True)

        # сохраняем по ширине колонок
#         for column in wb['Свод по муниципалитетам'].columns:
#             max_length = 0
#             column_name = get_column_letter(column[0].column)
#             for cell in column:
#                 try:
#                     if len(str(cell.value)) > max_length:
#                         max_length = len(cell.value)
#                 except:
#                     pass
#             adjusted_width = (max_length + 2)
#             wb['Свод по муниципалитетам'].column_dimensions[column_name].width = adjusted_width
        
        
        # Сохраняем с названием
        # создаем папку с датой
        
        wb.save(f'{path_end_folder}/Отчет {name_file} от {formatted_date}.xlsx')
        
        
        



        
        
```
